### **2D list and Its implementation using `NumPy`** 

#### *Manual implementation using `list`:*

Here we are implementing a matrix using a python 2D list. The rules are same as we did in 1D list but the rules of matrix operation is something that we have to implement. We are going to define three different function for addition , subtraction and multiplication. 
The matrices can be added if dimension of the matrices are similar otherwise tghe addition or subtraction doesnt hold any sense. The addition or subtraction is defined by
$$ A \pm B = C $$
$$ a_{ij} \pm b_{ij} = c_{ij} $$

However the multiplication of the matrix is defined by $C = AB \neq BA$ in general elementwise the same thing can be represented as $$ c_{ik} = a_{ij}b_{jk} $$
Here the repeated indices are summed over. Thus the matrix multiplication is only possible if coloumn number of the first matrix with the row number of the 2nd matrix. Mathematically if $A$ has dimension $(m\times n)$ then $B$ should have dimension $(n\times p)$ and their product will give us : $C_{m\times p}$

And also there is a formatting function `print_matrix(<matri>)` that does nothing but print the matrix in a more common way that we know

````python
# Create two 2D lists which will be used as matrix

A = [[1,0,0],[0,1,0],[0,0,1]]
B = [[1,3],[2,9],[5,8]]
M = [[1,3,9],[2,9,5],[5,8,7]]
````

````python
# Defining the functions to implement matrix operation

def print_matrix(A):
    for i in range(len(A)): # The len(A) gives us the number of rows
        for j in range(len(A[i])): # The len(A[0]) gives us the number of rows
            print(A[i][j],end = " ")
        print()
    print()

def mat_sum(A,B):
    row_A , coloumn_A = len(A),len(A[0])
    row_B , coloumn_B = len(B),len(B[0])
    if row_A != row_B :
        raise Exception("row mismatch error")
    elif coloumn_A != coloumn_B:
        raise Exception("coloumn Mismatch error") 
    else:
        C = [[0 for i in range(len(A[0]))] for j in range(len(A))]
        for i in range(row_A):
            for j in range(coloumn_A):
                C[i][j] = A[i][j] + B[i][j]
        return C

def mat_sub(A,B):
    row_A , coloumn_A = len(A),len(A[0])
    row_B , coloumn_B = len(B),len(B[0])
    if row_A != row_B :
        raise Exception("row mismatch error")
    elif coloumn_A != coloumn_B:
        raise Exception("coloumn Mismatch error") 
    else:
        C = [[0 for i in range(len(A[0]))] for j in range(len(A))]
        for i in range(row_A):
            for j in range(coloumn_A):
                C[i][j] = A[i][j] - B[i][j]
        return C

def mat_mul(A,B) -> list:
    row_A , coloumn_A = len(A),len(A[0])
    row_B , coloumn_B = len(B),len(B[0])
    if coloumn_A != row_B :
        raise Exception("Dimension mismatch error") # Throws error if there is a dimensional mismatch
    else:
        C = [[0 for i in range(coloumn_B)] for j in range(row_A)] # creates the appropiate dimensional place holder for upcoming data
        for i in range(row_A):
            for j in range(coloumn_A):
                for k in range(coloumn_B):
                    C[i][k] += A[i][j]*B[j][k] # the matrix multiplication 

        return C

````

````python
print_matrix(mat_mul(A,B))
print_matrix(mat_sum(A,M))
print_matrix(mat_sub(A,M))
````
