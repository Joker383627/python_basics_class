```python
a.append(89) # adds 89 at the end of the list
print(a)
b.clear() # clears the whole list
print(b)
d = a.copy() # copy the whole list a into a different variable
print(d)
print(a.count(4)) # counts the number of occurance of the number 4 in the list returns 0 if the element is not in the list
a.extend([78,23,44]) # add all the numbers to the list
print(a)
a.insert(2,6) # adds number 6 to the location before 2nd index... (counted from 0,1,2)
print(a)
poped_element = a.pop(3) # remove the element of index number 3 and returns it
print(a)
a.remove(9) # removes the element 9 from the list
print(a)
a.reverse() # reverse the array
print(a)
a.sort() # sorts the array
print(a)
```
